```python
list1 = [[9,80,205,40],[90,-60,96,1],[210,-3,101,89]]
for r in range(len(list1)):
    c = list1[r].index(max(list1[r]))
    k = 0
    while k < len(list1):
        if list1[r][c] <= list1[k][c]:
            k += 1
            if k == len(list1):
                print('鞍点在第{}行，第{}列，值为：{}'.format(r + 1, c + 1, list1[r][c]))
        else:
            break
```

```python
import numpy as np
import matplotlib.pyplot as plt
def fun(k):
    x = np.linspace(0,2*np.pi,1000)
    y = k*pow(x,2)*np.sin(x) + 2*k + np.cos(pow(x,3))
    return y
x = np.linspace(0,2*np.pi,1000)
color = 'bcgkmry'
lstyle = ['-','--','-.',':','--',':','-']
for i in range(1,7):
    y = fun(i)
    c = color[i]
    ls = lstyle[i]
    plt.plot(x,y,c=c,ls=ls)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def fun(k):
    x = np.linspace(0,2*np.pi,1000)
    y = k*pow(x,2)*np.sin(x) + 2*k + np.cos(pow(x,3))
    return y
x = np.linspace(0,2*np.pi,1000)
color = 'bcgkmry'
lstyle = ['-','--','-.',':','--',':','-']
for i in range(1,7):
    y = fun(i)
    c = color[i]
    ls = lstyle[i]
    plt.subplot(2,3,i)
    plt.plot(x,y,c=c,ls=ls)
    plt.xlabel('x')
    plt.ylabel('y')
plt.show()
```

```python
#z = x^2+y^2
import numpy as np
import  matplotlib.pyplot as plt
u = np.linspace(0,1,1000)
t = np.linspace(0,2*np.pi,1000)
U,T = np.meshgrid(u,t)
x = U*np.cos(T)
y = U*np.sin(T)
z = U*U
ax = plt.subplot(1,1,1,projection = '3d')
ax.plot_surface(x,y,z,cmap='viridis')
plt.show()
```

```python
import numpy as np
import  matplotlib.pyplot as plt
t = np.linspace(0,2*np.pi,1000)
s = np.linspace(-1,1,1000)
T,S = np.meshgrid(t,s)
x = (2+S*np.cos(T/2)/2)*np.cos(T)
y = (2+S*np.cos(T/2)/2)*np.sin(T)
z = S*np.sin(T/2)/2
ax = plt.subplot(1,1,1,projection = '3d')
ax.plot_surface(x,y,z,cmap='viridis')
plt.show()
```
