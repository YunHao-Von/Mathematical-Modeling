```python
import numpy as np
a = np.loadtxt('xiti1.txt')
R = a.copy()
for j in [0,2,3,5]:
    R[:,j] = R[:,j]/max(R[:,j])
for j in [1,4]:
    R[:,j] = min(R[:,j])/R[:,3]
np.savetxt('xiti1_1.txt',R)
```

```python
import numpy as np
from scipy.stats import rankdata
a = np.loadtxt('xiti1_1.txt')
cplus = a.max(axis=0)
cminus = a.min(axis=0)
print('正理想解=',cplus,'负理想解=',cminus)
d1 = np.linalg.norm(a-cplus,axis=1)
d2 = np.linalg.norm(a-cminus,axis=1)
print(d1,d2)
f1 = d2/(d1+d2)
print('TOPSIS的评价值为：',f1)

t = cplus-a
mmin = t.min()
mmax = t.max()
rho = 0.5
xs = (mmin+rho*mmax)/(t+rho*mmax)
f2 = xs.mean(axis = 1)
print('\n关联系数=',xs,'\n关联度=',f2)

[n,m] = a.shape
cs = a.sum(axis=0)
p = a/cs
e = -(p*np.log(p)).sum(axis=0)/np.log(n)
g = 1-e
w = g/sum(g)
F = p@w
print('\nP={}\n,e={}\n,g={}\n,w={}\nF={}'.format(p,e,g,w,F))

R = [rankdata(a[:,i]) for i in np.arange(6)]
R = np.array(R).T
print('\n秩矩阵为：\n',R)
RSR = R.mean(axis=1)/n
print('RSR=',RSR)
```
