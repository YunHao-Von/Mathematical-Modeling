```python
import numpy as np
from scipy.interpolate import interp1d
x = np.linspace(0,24,49)
y = np.array([2,0,0,0,2,0,0,0,0,0,2,0,5,0,8,0,25,0,12,0,0,5,0,10,0,12,
              0,0,7,0,0,0,9,0,28,0,22,0,10,0,9,0,11,0,8,0,9,0,3])
xnew = np.linspace(0,24,24*60) #插值点
f = interp1d(x,y)  #分段线性插值
y1 = f(xnew)
print('预计一天内有{}辆车通过'.format(int(sum(y1))))
f2 = interp1d(x,y,'cubic')
y2 = f2(xnew)
print('预计一天内有{}辆车通过'.format(int(sum(y2))))  #三次样条插值
```

```python
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
t = np.linspace(1,15,15)
x = t*6
y = np.array([352,211,197,160,142,106,104,60,56,38,36,32,21,19,15])
f = interp1d(x,y)
f1 = interp1d(x,y,'cubic')
xnew = np.linspace(6,15*6,15*6)
y1 = f(xnew)
y2 = f1(xnew)
plt.plot(x,y,'ro',xnew,y1,'b-',xnew,y2,'g--')
plt.xlabel('min')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy.integrate import quad
x = np.linspace(0,1,1000)
f = lambda x: (x+2)/(np.sqrt(x+1))
y = f(x)
f2 = interp1d(x,y,'cubic')
xnew = np.linspace(0,1,1000)
y2 = f2(xnew)
plt.rc('font',size = 16)
plt.rc('text',usetex=True)
plt.rc('font',family = 'SimHei')
plt.plot(x,y,'r^',label='A')
plt.plot(xnew,y2,'g-',label = 'B')
plt.legend()
plt.show()

jifen1 = quad(f,0,1)
jifen2 = quad(f2,0,1)
print('积分1为：',jifen1)
print('积分2为：',jifen2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
x = np.array([0.460556,1.38208,2.39639,3.41056,4.42472,5.43903,
              6.45319,7.4675,8.44819,11.4935,12.4935,13.4151,
              14.429,15.4431,16.365,17.3789,18.4846,19.4985,
              20.3993,24.4335,25.4476])
y = np.array([13.471,11.5953,10.3498,9.73471,9.48735,8.69649,
              9.4897,8.90086,10.1036,18.5833,19.6766,18.6466,
              16.0463,16.5697,15.5428,14.677,14.6733,15.5294,
              15.1898,13.4514,11.8095])
f = np.polyfit(x,y,8)
f1 = interp1d(x,y,'cubic')
xnew = np.linspace(1,25,100)
y1  = f1(xnew)
print('拟合多项式系数：',f)
plt.rc('font',size=16)
plt.rc('font',family = 'SimHei')
plt.plot(x,y,'*',x,np.polyval(f,x),'-',xnew,y1,'--')
plt.legend(['point','nihe','chazhi'])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = 1960+ np.linspace(0,8,9)
y = np.array([2972,3061,3151,3213,3234,3285,3356,3420,3483])
plt.rc('font',size=16)
plt.plot(x,y,'*')
for i in range(1,5):
    f = np.polyfit(x,y,i)
    print('拟合多项式系数为：',f)
    plt.plot(x,np.polyval(f,x),'-')

plt.legend(['0','1','2','3','4'])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
x = np.arange(1,11.0,1.0)
y = np.array([2615.0,1943.0,1494.0,1087.0,765.0,538.0,484.0,290.0,226.0,204.0])
y1 = lambda x,a,b,c: 1/(a*x**2+b*x+c)
plt.rc('font',size=16)
plt.plot(x,y,'^')
popt,pcov = curve_fit(y1,x,y)
print('拟合的参数值为：',popt)
y2 = y1(x,*popt)
plt.plot(x,y2,'-')
print('预测4.5年后的轿车的平均价格为：',y1(4.5,*popt))
plt.show()
```

```python
import numpy as np
from scipy.interpolate import interp1d
from matplotlib.pyplot import plot,rc,show
x = np.array([7.0,10.5,13.0,17.5,34.0,40.5,44.5,48.0,56.0,
              61.0,68.5,76.5,80.5,91.0,96.0,101.0,104.0,106.5,
              111.5,118.0,123.5,136.5,142.0,146.0,150.0,157.0,158.0])
y1 = np.array([44,45,47,50,50,38,30,30,34,
               36,34,41,45,46,43,37,33,28,
               32,65,55,54,52,50,66,66,68
               ])
y2 = np.array([44,59,70,72,93,100,110,110,110,
               117,118,116,118,118,121,124,121,121,
               121,122,116,83,81,82,86,85,68])
f1 = interp1d(x,y1,'linear')
f2 = interp1d(x,y2,'linear')
newx = np.arange(7,158,0.1)
y11 = f1(newx)
y22 = f2(newx)
rc('font',size=16)
plot(newx,y11,'-',newx,y22,'-')
show()
area = np.sum(y22-y11)*1600*0.1/18**2
print('估计的面积为：',area,'km^2')
```
