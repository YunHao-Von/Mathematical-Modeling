```python
import numpy as np
import numpy.linalg as la
mu = np.array([0,0]).reshape(2,1)
su = np.array([[1,0.9],[0.9,1]])
A = np.array([1,1]).reshape(2,1)
B = np.array([1,-1]).reshape(2,1)
#马氏距离
mahalanobis = np.sqrt((A-B).T@la.inv(su)@(A-B))
print('马氏距离结果为：',mahalanobis[0])
#欧氏距离
sum = 0
for i in range(len(A)):
    sum = sum + (A[i]-B[i])**2
euclidean = np.sqrt(sum)
print('欧氏距离结果为：',euclidean)
#曼哈顿距离
total = 0
for i in range(len(A)):
    total = total + np.abs(A[i]-B[i])
manhattan = total
print('曼哈顿距离结果为：',manhattan)
#切比雪夫距离
maxmanhattan = []
for i in range(len(A)):
    d = np.abs(A[i]-B[i])
    maxmanhattan.append(d)
chebyshev = np.max(maxmanhattan)
print('切比雪夫距离结果为：',chebyshev)
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
x1 = np.array([[3,2],[2,4],[4,7]])
x2 = np.array([[6,9],[5,7],[4,8]])
m1 = np.mean(x1,axis=0)
m2 = np.mean(x2,axis=0)
print(m1,'\n\n',m2,'\n')
v1 = np.cov(x1.T)
v2 = np.cov(x2.T)
print(v1,'\n\n',v2,'\n')
x0 = np.r_[x1,x2]
v = np.cov(x0.T)
y0 = np.hstack([np.ones(3),2*np.ones(3)])
#判别
knn = KNeighborsClassifier(2,metric='mahalanobis',metric_params={'V':v})
knn.fit(x0,y0)

#预测
x = np.array([[2,7]])
pre = knn.predict(x)
print('马氏距离分类结果：',pre)
print('马氏距离已知样本的误判率为：',1-knn.score(x0,y0))
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
a = np.loadtxt('11_14.txt')
x0 = a[0:-3,1:]
y0 = np.hstack([np.ones(3),2*np.ones(5)])
x = a[-3:,1:]
v = np.cov(x0.T)
knn = KNeighborsClassifier(2,metric='mahalanobis',metric_params={'V':v})
knn.fit(x0,y0)
pre = knn.predict(x)
print('马氏距离分类结果：',pre)
print('马氏距离已知样本的误判率为：',1-knn.score(x0,y0))
print('交叉误判率为：',cross_val_score(knn,x0,y0,cv=2))
print('\n')
knn2 = KNeighborsClassifier(2)
knn2.fit(x0,y0)
pre2 = knn2.predict(x)
print('欧氏距离分类结果：',pre2)
print('欧氏距离已知样本的误判率为：',1-knn2.score(x0,y0))
print('交叉误判率为：',cross_val_score(knn2,x0,y0,cv=2))
print('\n')
clf = LDA()
clf.fit(x0,y0)
print('Fisher准则判别结果为：',clf.predict(x))
print('Fisher准则已知样本的误判率为：',1-clf.score(x0,y0))
model = LDA()
print('交叉误判率为：',cross_val_score(model,x0,y0,cv=2))
print('\n')
clf2 = GaussianNB()
clf2.fit(x0,y0)
print('贝叶斯判别法结果为：',clf.predict(x))
print('贝叶斯判别法已知样本的误判率为：',1-clf.score(x0,y0))
model1 = GaussianNB()
print('交叉误判率为：',cross_val_score(model1,x0,y0,cv=2))
```

```python
import numpy as np
from sklearn.decomposition import PCA
a = np.loadtxt('11_15.txt')
md = PCA().fit(a) #构造并训练模型
print('特征值：',md.explained_variance_)
print('各主成分的贡献率：',md.explained_variance_ratio_)
print('奇异值为：',md.singular_values_)
print('各主成分系数为：\n',md.components_)
```

```python
import numpy as np
from scipy.stats import zscore
a = np.loadtxt('11_15.txt')
print('相关系数矩阵为：\n',np.corrcoef(a.T))
# b = np.delete(a,0,axis=1) #删除第一列数据
c = zscore(a) #标准化
r = np.corrcoef(c.T)  #标准化相关系数矩阵
d,e = np.linalg.eig(r)  #特征值和特征向量
rate = d/d.sum()  #计算各主成分的贡献率
print('特征值为：',d)
print('特征向量为：\n',e)
print('各主成分的贡献率为：',rate)
k = 1
F = e[:,:k]
score_mat = c.dot(F)
score1 = score_mat.dot(rate[0:k])
score2=-score1
print('各评价对象的得分为：',score2)
index = score1.argsort()+1
print('从高到低各个城市的编号排序为：',index)
```
