```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
def pfun(y,t):
    i,s=y
    return np.array([s*i-0.3*i,-s*i])
t=np.arange(0,50,0.01)
sol1=odeint(pfun,[0.02,0.98],t)
plt.rc('font',size=16)
plt.rc('font',family='SimHei')
plt.grid(True)
plt.plot(t,sol1[:,0],'r*',label='i(t)')
plt.plot(t,sol1[:,1],'r*',label='s(t)')
plt.figure(2)
plt.grid(True)
plt.plot(sol1[:,1],sol1[:,0],'b-',label='i-t')
```
