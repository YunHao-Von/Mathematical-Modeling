```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
def pfun(x,t):
    x1=x
    return np.array([0.03*x1*(1-x1/30000)])
t=np.arange(0,10,0.1)
sol1=odeint(pfun,[100],t)
```

```python
from scipy.integrate import odeint
from numpy import arange
import matplotlib.pyplot as plt
dx=lambda x,t:0.03*x*(1-x/30000)
t=arange(0,350,0.1)
sol=odeint(dx,100,t)
plt.plot(t,sol[:,0],'r-',label="logistic_model")
plt.legend()
plt.show()
```

```python
from sympy.abc import x
from sympy import diff, symbols,dsolve, simplify, Function
y=Function('y')
r=symbols('r')
xm=symbols('xm')
x0=symbols('x0')
eq=diff(y(x),x)-r*y(x)+r/xm*y(x)*y(x)  #定义方程
con={y(0):x0}
result=dsolve(eq,ics=con)
print(simplify(result))
```

```python
from sympy.abc import t
from sympy import diff, symbols,dsolve, simplify, Function
x=Function('x')
a=symbols('a')
x0=symbols('x0')
eq=diff(x(t),t)-a*x(t)+a*x0
result=dsolve(eq)
print(sp.simplify(result))
```
