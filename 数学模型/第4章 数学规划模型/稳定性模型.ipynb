```python
#捕鱼业的持续收获
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
r,x,N,E=symbols('r,x,N,E')
print(solve(r*x*(1-x/N)-E*x,x))
x=np.arange(0,100000,0.1)
N=100000
r=0.05
E=0.01
y=r*x*(1-x/N)
plt.grid(True)
plt.plot(x,y,'b-',label='$y=f(x)$')
t=np.arange(0,30000,0.1)
plt.plot(t,r*t,'r--',label='$y=rx$')
plt.legend()
```
