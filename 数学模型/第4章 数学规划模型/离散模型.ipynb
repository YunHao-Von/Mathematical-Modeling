```python
def count_probably(k):
    fenbu=binom(300,0.05)
    if(k==0):
        return fenbu.cdf(0)
    else:
        return fenbu.cdf(k)-fenbu.cdf(k-1)
def count_j(m):
    xigema=0
    for k in range(0,m-300):
        xigema=xigema+(m-k-300)*count_probably(k)
    jm=1/(0.6*300)*(0.95*m-(1+0.2)*xigema)-1
    return jm
```

```python
lamda=0.6
n=300
bg=0.2
m=np.arange(300,330,2)
j=[]
for i in range(300,330,2):
    j.append(count_j(i))
import matplotlib.pyplot as plt
plt.grid(True)
plt.plot(m,j,'r--',label='count_j')
plt.legend()
plt.text(310,0.63,"$\\leftarrow\\alpha$"+'=0.1')
plt.show()
```
